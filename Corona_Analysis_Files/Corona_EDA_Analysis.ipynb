```python
Covid_Data = pd.read_csv('F:/ML_DataScience_ImageProcessing/DataSets/Corona_Virus_Dataset/covid_19_data.csv')
```

```python
Covid_Data.describe()
```

```python
Covid_Data.head()
```

## Country Wise Confirmed Report Found 

```python
plt.figure(figsize=(45,15))
chart = sns.countplot(
    data=Covid_Data[Covid_Data['Confirmed'] != 0],
    x='Country/Region',
    palette='Set1'
)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
for p in chart.patches:
    height = p.get_height()
    chart.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{}'.format(height),
            ha="center")
    
plt.show()
```

## Country Wise Recovery Cases Found

```python

plt.figure(figsize=(25,15))
chart = sns.countplot(
    data=Covid_Data[Covid_Data['Recovered'] != 0],
    x='Country/Region',
    palette='Set1'
)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
for p in chart.patches:
    height = p.get_height()
    chart.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{}'.format(height),
            ha="center")
    
plt.show()
```
