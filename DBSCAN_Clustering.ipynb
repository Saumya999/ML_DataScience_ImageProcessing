## Importing the Libraries

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
%matplotlib inline
```

## Loading the Datastes 

```python
X, label = make_moons(n_samples= 2000, noise=0.1)
X[0:5]
```

```python
label
```

## Building Our Model 

```python
model = DBSCAN(eps=0.1,min_samples=10)
model.fit(X)
```

## My Prediction

```python
fig, ax = plt.subplots(figsize=(6,5))
ax.scatter(X[:,0], X[:,1], c=model.labels_)
fig.show()
```

## Original Prediction 

```python
fig, ax = plt.subplots(figsize=(6,5))
ax.scatter(X[:,0], X[:,1], c=label)
fig.show()
```
